```
import random, time
meyveler = ['şeftali', 'mango', 'elma', 'muz', 'kayısı', 'pineapple','çilek', 'grefut','avukado']
Sebzeler = ['pırasa', 'patates', 'patlıcan', 'kereviz', 'ıspanak', 'enginar']
userGuesslist = []
userGuesses = []
playGame = True
category = ""
continueGame = "D"

name = input("Lüften Adınızı Girin: ")
print("Merhaba", name.capitalize(), "Hadi Adam Asmaca Oynalım!")
time.sleep(1)


while True:    
    while True:
        if category.upper() == 'S':
            secretWord = random.choice(Sebzeler)
            break
        elif category.upper() == 'M':
            secretWord = random.choice(meyveler)
            break
        else:
            category = input("Lüften Kelime Kategorisi Seçin : M Meyveler / S Sebzeler ; Çıkış için X ")

        if category.upper() == 'X':
            print("Güle Güle. Yine Gel!")
            playGame = False
            break

    if playGame:
        secretWordList = list(secretWord)
        attempts = (len(secretWord) + 2)

        
        def printGuessedLetter():
            print("Gizli Kelimen : " + ''.join(userGuesslist))


    
        for n in secretWordList:
            userGuesslist.append('_')
        printGuessedLetter()

        print("Tahmin Hakkın :", attempts)


        #starting
        while True:

            print("Bir Harf Tahmin Et:")
            letter = input()

            if letter in userGuesses:
                print("Bu Harfi Zaten Denedin, Başka dene.")

            else:
                attempts -= 1
                userGuesses.append(letter)
                if letter in secretWordList:
                    print("Harika Tahmin!")
                    if attempts > 0:
                        print( attempts, 'şansın kaldı!')
                    for i in range(len(secretWordList)):
                        if letter == secretWordList[i]:
                            letterIndex = i
                            userGuesslist[letterIndex] = letter.upper()
                    printGuessedLetter()

                else:
                    print("Olmadı! Tekrar Dene.")
                    if attempts > 0:
                        print( attempts, 'Şansın Kaldı!')
                    printGuessedLetter()


            #Win/loss
            joinedList = ''.join(userGuesslist)
            if joinedList.upper() == secretWord.upper():
                print("Tebrikler Kazandın!")
                break
            elif attempts == 0:
                print(" Şansın Kalmadı!, Başka Sefere artık.")
                print("Gizli Kelime: "+ secretWord.upper())
                break

        #Play again
        continueGame = input("Tekrar Oynamak İster Misin ? D Devam İçin, X Çıkış için ")
        if continueGame.upper() == 'D':
            category = input("Lüften Kelime Kategorisi Seçin : M Meyveler / S Süper Kahraman")
            userGuesslist = []
            userGuesses = []
            playGame = True
        else:
            print("Oyun İçin Teşekkürler. Bye Bye!")
            break
    else:
        break
```
